## 접근한 웹 페이지에서 원하는 데이터 추출하고 정리하기

```python
tmp_one = soup.find_all('div', 'sammy')[0]
type(tmp_one)
```

```python
tmp_one.find(class_ = 'sammyRank')
```

```python
tmp_one.find(class_ = 'sammyRank').get_text()
```

```python
tmp_one.find(class_ = 'sammyListing').get_text()
```

```python
tmp_one.find('a')['href']
```

```python
import re

tmp_string = tmp_one.find(class_ = 'sammyListing').get_text()

re.split(('\n|\r\n'), tmp_string)

print(re.split(('\n|\r\n'), tmp_string)[0])
print(re.split(('\n|\r\n'), tmp_string)[1])
```

```python
from urllib.parse import urljoin
```

```python
rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div', 'sammy')

for item in list_soup :
    rank.append(item.find(class_ = 'sammyRank').get_text())
    
    tmp_string = item.find(class_ = 'sammyListing').get_text()
    
    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

```python
rank[:5]
```

```python
main_menu[:5]
```

```python
cafe_name[:5]
```

```python
url_add[:5]
```

```python
len(rank), len(main_menu), len(cafe_name), len(url_add)
```

```python
import pandas as pd

data = {'Rank' : rank, 'Menu' : main_menu, 'Cafe' : cafe_name, 'URL' : url_add}
df = pd.DataFrame(data)
df.head()
```

```python
df = pd.DataFrame(data, columns = ['Rank', 'Cafe', 'Menu', 'URL'])
df.head(5)
```

```python
df.to_csv('../data/03.best_sandwiches_list_chicago.csv', sep = ',', encoding = 'UTF-8')
```

## 다수의 웹 페이지에 자동으로 접근해서 원하는 정보 가져오기

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

import pandas as pd
```

```python
df = pd.read_csv('../data/03.best_sandwiches_list_chicago.csv', index_col = 0)
df.head()
```

```python
df['URL'][0]
```

```python
html = urlopen(df['URL'][0])
soup_tmp = BeautifulSoup(html, 'html.parser')
soup_tmp
```

```python
print(soup_tmp.find('p', 'addy'))
```

```python
price = []
address = []

for n in df.index[:3] :
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'lxml')
    
    gettings = soup_tmp.find('p', 'addy').get_text()
    
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
```

## Jupyter Notebook에서 상태 진행바를 쉽게 만들어주는 tqdm 모듈
## 상태 진행바 적용 후 다시 샌드위치 페이지 50개에 접근하기

```python
from tqdm import tqdm_notebook

price = []
address = []

for n in tqdm_notebook(df.index) :
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'lxml')
    
    gettings = soup_tmp.find('p', 'addy').get_text()
    
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
```

```python
price
```

```python
address
```

```python
len(price), len(address), len(df)
```

```python
df['Price'] = price
df['Address'] = address

df = df.loc[:, ['Rank', 'Cafe', 'Menu', 'Price', 'Address']]
df.set_index('Rank', inplace = True)
df.head()
```

```python
df.to_csv('../data/03.best_sandwiches_list_chicago2.csv', 
          sep = ',', encoding = 'UTF-8')
```

## 맛집 위치를 지도에 표기하기

```python
import folium
import pandas as pd
import googlemaps
import numpy as np
```

```python
df = pd.read_csv('../data/03.best_sandwiches_list_chicago2.csv',
                index_col = 0)
df.head(5)
```

```python
gmaps_key = '********'
gmaps = googlemaps.Client(key = gmaps_key)
```

```python
lat = []
lng = []

for n in tqdm_notebook(df.index) :
    if df['Address'][n] != 'Multiple' :
        target_name = df['Address'][n] + ',' + 'Cicago'
        
        gmaps_output = gmaps.geocode(target_name)
        
        location_output = gmaps_output[0].get('geometry')
        
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else :
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
df['lat'] = lat
df['lng'] = lng
df.head()
```

```python
mapping = folium.Map(location = [df['lat'].mean(), df['lng'].mean()],
                     zoom_start = 11)

folium.Marker([df['lat'].mean(), df['lng'].mean()], 
              popup = 'center').add_to(mapping)
        
mapping
```

```python
mapping = folium.Map(location = [df['lat'].mean(), df['lng'].mean()],
                     zoom_start = 11)

for n in df.index :
    if df['Address'][n] != 'Multiple' :
        folium.Marker([df['lat'][n], df['lng'][n]],
                      popup = df['Cafe'][n]).add_to(mapping)
        
mapping
```
